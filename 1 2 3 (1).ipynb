```python
mode = dataset["Country"].mode()[0] #most frequent value
print(mode)
dataset["Country"].fillna(mode, inplace=True)
print(dataset)
```

```python
import matplotlib.pyplot as plt


state = data["state"]
gender = data["gender"]

state_cut = state[:20]
gender_cut = gender[:20]

gender_count = gender_cut.value_counts()

plt.figure(figsize=(10,6))
plt.bar(gender_count.index, gender_count.values, color='red')
plt.title('Gender Distribution (First 20 Records)')
plt.xlabel('Gender')
plt.ylabel('Count')
#plt.xticks(rotation=45)
plt.show()
```

```python
# binning
num = pd.DataFrame({ 
    'num': np.random.randint(1,100,10)    
})

num['num_bins'] = pd.cut(x=num['num'], bins=[1,25,50,75,100])

print(num)
```

```python
#MIN MAX SCALER
data2=pd.read_csv('diamonds.csv')
min_max_scaler = preprocessing.MinMaxScaler()
# Create an object to transform the data to fit minmax processor
#dataset diamond
x_scaled = min_max_scaler.fit_transform(data2[['price']].values.astype(float))
# Run the normalizer on the dataframe
df_normalized = pd.DataFrame(x_scaled)
data2['normalized2']=x_scaled
data2.sort_values(by='price')
```

```python
#ZSCORE
from scipy import stats
data2['zscore']=stats.zscore(data2['price'])
data2.sort_values(by='price')
```
